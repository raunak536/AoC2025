```python
def find_next_strike(x, y, grid):
    "Find next strike on ^ of if end of grid is found"
    x_next, y_next = x + 1, y
    while x_next <= (len(grid)-1):
        if grid[x_next][y_next] == '.':
            x_next, y_next = x_next + 1, y_next
            continue
        else:
            # print(x_next, y_next)
            return '^',(x_next, y_next)
    return 'passed',(x_next-1, y_next)
```

```python
visited = set()
def fire_beam(x_start, y_start, grid):
    global visited
    if not (0 <= y_start < len(grid[0])):
        return 0
    status_strike, (x_strike, y_strike) = find_next_strike(x_start, y_start, grid)
    if status_strike=='passed':
        return 0
    else:
        count = 0
        if (x_strike, y_strike) not in visited:
            visited.add((x_strike, y_strike))
            count = 1
            count += fire_beam(x_strike, y_strike-1, grid)
            count += fire_beam(x_strike, y_strike+1, grid)
    return count
```

```python
visited = set()
fire_beam(0,7,test_grid)
```

```python
visited = set()
fire_beam(0, 70, grid)
```
