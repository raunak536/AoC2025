### Part 1 AND 2

```python
def max_joltage(digit_string):
    max_digit = -9
    max_digit_idx = None
    for i in range(len(digit_string)):
        if int(digit_string[i]) > max_digit:
            max_digit = int(digit_string[i])
            max_digit_idx = i
    
    return max_digit, max_digit_idx
```

```python
def total_joltage(inputs, batteries_cnt=2):
    total_joltage = 0
    for input_string in inputs:
        start_idx = 0
        end_idx=-batteries_cnt+1
        joltage = ''
        # print(input_string)
        for cnt in range(1,batteries_cnt+1):
            # print(start_idx, end_idx)
            digit, digit_idx = max_joltage(input_string[start_idx:end_idx])
            start_idx=digit_idx+1
            end_idx=-batteries_cnt+1+cnt
            if end_idx==0:
                end_idx=len(input_string)
            joltage+=str(digit)
        # print(joltage)
        total_joltage += int(joltage)
    return total_joltage
```

```python
total_joltage(inputs, 2)
```
