#### Part 1

```python
import numpy as np
arr = np.array([list(l) for l in inputs],dtype=str)
```

```python
def count_adjacent_papers(arr, x, y):
    count = 0
    visited_coordinates = [(x,y)]
    for i in range(-1,2):
        for j in range(-1,2):
            try:
                x_adj = max(0,x+i)
                y_adj = max(0,y+j)
                if (x_adj, y_adj) in visited_coordinates:
                    continue
                # print(x_adj,y_adj)
                adjacent_coordinate = arr[x_adj, y_adj]
                visited_coordinates.append((x_adj, y_adj))
                if adjacent_coordinate == '@':
                    # print(x_adj,y_adj)
                    count+=1
            except IndexError:
                pass
    return count

count_adjacent_papers(arr,1,1)
```

```python
total_rolls = (arr == '@').sum()
accessible_coordinates = 0
for x_coordiante in range(140):
    for y_coordiante in range(140):
        if arr[x_coordiante, y_coordiante] == '@':
            cnt = count_adjacent_papers(arr, x_coordiante, y_coordiante)
            if cnt<4:
                accessible_coordinates+=1
print(f"Total paper rolls which are accessible : {accessible_coordinates} | Out of possible rolls : {round(100*accessible_coordinates/total_rolls,1)}% ")
```

### Part 2

```python
possible_to_remove_rolls = True
removed_rolls = 0
iterations = 0
while possible_to_remove_rolls:
    iterations+=1
    possible_to_remove_rolls = False
    for x_coordiante in range(140):
        for y_coordiante in range(140):
            if arr[x_coordiante, y_coordiante] == '@':
                cnt = count_adjacent_papers(arr, x_coordiante, y_coordiante)
                if cnt<4:
                    arr[x_coordiante, y_coordiante] = '.'
                    possible_to_remove_rolls = True
                    removed_rolls+=1


print(f"Total paper rolls which were removed : {removed_rolls} in {iterations} iterations | Out of possible rolls : {round(100*removed_rolls/total_rolls,1)}% ")
```
