```python
import numpy as np
```

```python
inputs = open('inputs.txt','r').read().splitlines()
```

```python
len(inputs)
```

### Part 1

```python
inputs_arr = [i.split() for i in inputs[:-1]]
inputs_arr = np.array(inputs_arr, dtype='int')
operand_arr = np.array(inputs[-1].split())

sums = inputs_arr.sum(axis=0)
multiples = inputs_arr.prod(axis=0)

sums[operand_arr == '+'].sum() + multiples[operand_arr == '*'].sum()
```

### Part 2

```python
grid = np.array([list(x.rstrip('\n')) for x in inputs], dtype='str')

grid_digits = grid[:-1]
grid_operands = grid[-1]

operand_indexes = []
for i in range(len(grid_operands)):
    if grid_operands[i] != ' ':
        operand_indexes.append(i)


ranges = []
for i in range(1,len(operand_indexes)):
    ranges.append(list(range(operand_indexes[i-1], operand_indexes[i]-1)))
ranges.append(list(range(operand_indexes[-1], len(grid_operands))))

new_arr = []
for r in ranges:
    col_arr = []
    for col_idx in r:
        d = ''
        for j in range(4):
            num = grid_digits[j][col_idx]
            # print(num, j, col_idx)
            d += num
        col_arr.append(int(d))
    new_arr.append(col_arr)

len(new_arr)

totals = 0
for e,j in enumerate([i for i in grid_operands if i!=' ']):
    if j == '+':
        totals += np.array(new_arr[e]).sum()
    else:
        totals += np.array(new_arr[e]).prod()
totals        
```
