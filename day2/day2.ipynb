### part 1

```python
def is_valid(id: str):
    N = len(id)
    if N%2 != 0:
        return False
    else:
        mid = N//2
        if id[:mid] == id[mid:]:
            return True
        else:
            return False
```

```python
invalid_ids = []
invalid_ids_sum = 0
for id_range in inputs.split(','):
    start, end = id_range.split('-')
    for id in range(int(start), int(end)+1):
        if is_valid(str(id)):
            invalid_ids.append(id)
            invalid_ids_sum+=int(id)
```

### part 2

```python
def is_valid_2(id: str):
    N = len(id)
    if N==1:
        return False
    for repeat_seq_len in range(1, 1 + N//2):
        invalid_set = set()
        for i in range(0, N, repeat_seq_len):
            invalid_set.add(id[i:i+repeat_seq_len])
        # print(repeat_seq_len, invalid_set)
        if len(invalid_set) == 1:
            return True
    return False
```

```python
invalid_ids = []
invalid_ids_sum = 0
for id_range in inputs.split(','):
    start, end = id_range.split('-')
    for id in range(int(start), int(end)+1):
        if is_valid_2(str(id)):
            invalid_ids.append(id)
            invalid_ids_sum+=int(id)
invalid_ids_sum
```
