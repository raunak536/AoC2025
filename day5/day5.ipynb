```python
i = 0
ID_range = inputs[i]
fresh_IDs = []
while ID_range!='':
    start, end = ID_range.split('-')
    fresh_IDs.append((int(start), int(end)))
    i+=1
    ID_range=inputs[i]

fresh_count = 0
for j in range(i+1, len(inputs)):
    available_ID = int(inputs[j])
    for start,end in fresh_IDs:
        if ((available_ID>=start) & (available_ID<=end)):
            fresh_count+=1
            break
            # print(available_ID, start, end)

print(f"Fresh ingredients found : {fresh_count} | Out of {round(100 * fresh_count/(len(inputs)-i) ,2)}%")
```

### Part 2

```python
len(fresh_IDs)
```

```python
unqiue_fresh_IDs = list(set(fresh_IDs))
unqiue_fresh_IDs  = sorted(unqiue_fresh_IDs, key = lambda x : x[0])
unqiue_fresh_IDs = [[i[0],i[1]] for i in unqiue_fresh_IDs]
merged = []
for interval in unqiue_fresh_IDs:
    if (len(merged)==0) or (merged[-1][1] < interval[0]):
        merged.append(interval)
    else:
        merged[-1][1] = max(merged[-1][1], interval[1])
```

```python
available_IDs = 0
for i in merged:
    available_IDs += i[1] - i[0] + 1
available_IDs
```
